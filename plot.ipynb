```python
legendgroup=[float(i) for i in data_sobol['all_runs']],
x = [data_sobol['all_runs'][i][0] for i in data_sobol['all_runs']]
y = [data_sobol['all_runs'][i][1] for i in data_sobol['all_runs']]
z = [data_sobol['all_runs'][i][2] for i in data_sobol['all_runs']]
fig = go.Figure(data=[go.Scatter3d(x=x, 
                                   y=y, 
                                   z=z,
                                   mode='markers',
                                   marker=dict(size=3,color=[float(i) for i in data_sobol['all_runs']], colorscale='Jet', reversescale=True, showscale=True)
                                  )])
fig.update_layout(scene = dict(
                    xaxis_title='Selenzyme',
                    yaxis_title='FBA',
                    zaxis_title='Thermodynamics'))

fig.update_layout(
    scene=dict(
        annotations=[dict(
            x=data_sobol['all_runs'][str(min([float(i) for i in data_sobol['all_runs']]))][0],
            y=data_sobol['all_runs'][str(min([float(i) for i in data_sobol['all_runs']]))][1],
            z=data_sobol['all_runs'][str(min([float(i) for i in data_sobol['all_runs']]))][2],
            text="Best: "+str(data_sobol['all_runs'][str(min([float(i) for i in data_sobol['all_runs']]))]),
            textangle=0,
            ax=0,
            ay=-75,
            font=dict(
                color="black",
                size=12
            ),
            arrowcolor="black",
            arrowsize=3,
            arrowwidth=1,
            arrowhead=1
        )]
    ),
)

fig.show()
```
